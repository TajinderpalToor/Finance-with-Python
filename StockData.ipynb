```python
#getting list of holdings in VUG ETF
import requests
import re

#using four etfs
#1. VUG - vanguard growth etf
#2. VIOG - Vanguard S&P Small Cap 600 Growth ETF
#3. ARKK - ARK Inovation ETF
#4. ARKF - Fintech Innovation ETF

keys = ['VUG', 'VIOG','ARKK','ARKF']


headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:83.0) Gecko/20100101 Firefox/83.0"
}


def main(url):
    with requests.Session() as req:
        req.headers.update(headers)
        for key in keys:
            r = req.get(url.format(key))
            print(f"Extracting: {r.url}")
            goal = re.findall(r'etf\\\/(.*?)\\', r.text)
            print(goal)


main("https://www.zacks.com/funds/etf/{}/holding")
```

```python
import seaborn as sns
from scipy.stats import shapiro
from statistics import mean
import numpy as np
def distributions(sector):

  instut_hold=[]
  fwdPE=[]
  ann_return=[]
  #keys
  keys_dict=[]

  for keys in sector:
    keys_dict.append(keys)
    instut_hold.append(sector[keys][0])
    fwdPE.append(sector[keys][1])
    ann_return.append(sector[keys][2])

  if len(keys_dict) < 20:
    holder=keys_dict
    print(len(holder))
    return holder

  else:
    #get highest 20 of each
    #print("The highest 20 holdings are:")
    instituion_holdings=(sorted(range(len(instut_hold)), key = lambda sub: instut_hold[sub])[-10:])
    #print("***************")


    # get mean of fwdPE and then make a temp array which divides the values by average to see how much bigger, choose the 20 bigger than this
    mean_PE=mean(fwdPE)
    temp=np.divide(fwdPE,mean_PE)
    #print("The highest 20 PE are:")
    PEratio=(sorted(range(len(temp)), key = lambda sub: temp[sub])[-10:])
    #print("***************")

    #print("The highest 20 returns are:")
    annualreturn=(sorted(range(len(ann_return)), key = lambda sub: ann_return[sub])[-10:])
    #print("***************")

    annre_tick=[]
    pe_tick=[]
    intut_tick=[]

    for i in range(-1,-10,-1):
      annre_tick.append(keys_dict[annualreturn[i]])
      pe_tick.append(keys_dict[PEratio[i]])
      intut_tick.append(keys_dict[instituion_holdings[i]])


    at=set(annre_tick)
    pt=set(pe_tick)
    ar=set(intut_tick)
    holder=at&pt&ar

    if len(holder)==0:
      holder_one=at&pt
      holder_one=list(holder_one)
      print(len(holder_one))
      return holder_one

    elif (len(at&pt))==0:
      holder_two=at
      holder_two=list(holder_two)
      print(len(holder_two))
      return holder_two
    else:
      holder=list(holder)
      print(len(holder))
      return holder


```

```python
# get the indexes we made, get the coresponding stocks from the keys_dict, find all the values that are present in all three lists keep those tickers
#do this for all the dictionaries

#CONVERT EACH SET TO STRING
t=distributions(Technology)
a=distributions(CommunicationServices)
b=distributions(BasicMaterials)
c=distributions(Energy)
d=distributions(RealEstate)
e=distributions(Healthcare)
f=distributions(ConsumerCyclical)
g=distributions(Industrials)

#no need to do utilities - just take all theres only 8
h=distributions(Utilities)

i=distributions(ConsumerDefensive)
j=distributions(FinancialServices)

```

```python
#combine the lists
FinalStock_tickers= t+a+b+c+d+e+f+g+h+i+j
print(len(FinalStock_tickers))
```
