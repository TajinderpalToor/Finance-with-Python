```python
import talib as ta
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
import datetime as dt
import yfinance as yf
```

```python
aapl=yf.Ticker("AAPL")
hist = aapl.history(period="5y")
```

```python
hist
```

```python
# calculate moving average -- average of last 100 days
hist['SMA_100']=ta.SMA(hist['Close'],100)
```

```python
plt.plot(hist['Close'])
plt.plot(hist['SMA_100'])
```

```python
#exponential moving average
hist['EMA_100']=ta.EMA(hist['Close'],100)
```

```python
plt.plot(hist['Close'])
plt.plot(hist['EMA_100'])
```

```python
#relative strength - momentum indicator (over 70%rsi = asset is over bought, less then 30 asset is oversold),default is 14 days
hist['RSI']=ta.RSI(hist['Close'])

fig,axs=plt.subplots(2,1,gridspec_kw={"height_ratios":[3,1]},figsize=(10,6))
axs[0].plot(hist['Close'])
axs[1].axhline(y=70,color="r",linestyle="--")
axs[1].axhline(y=30,color="r",linestyle="--")
axs[1].plot(hist['RSI'],color="orange")
```

```python
#MACD - moving average convergence and diveregence
end_date = dt.datetime.now()

gs=yf.Ticker('GS')
data=gs.history(start=dt.datetime(2021,1,1),end=end_date)
macd,macd_signal,macd_hist=ta.MACD(data['Close'])

fig,axs=plt.subplots(2,1,gridspec_kw={"height_ratios":[3,1]},figsize=(10,6))
c=["red" if cl< 0 else "green" for cl in macd_hist]
axs[0].plot(data['Close'])
axs[1].plot(macd,'b-')
axs[1].plot(macd_signal,'--',color="orange")
axs[1].bar(macd_hist.index,macd_hist,color=c)
```
